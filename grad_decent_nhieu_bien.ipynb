```python
import numpy as np 


def grad_x(x, y):
    return 4*x*(x*x + y - 7) + 2*(x - y + 1)

def grad_y(x, y):
    return 2*(x*x + y -7) - 2*(x - y + 1)

def grad(w):
    x = w[0][0]
    y = w[1][0]
#     print(grad_x(x,y), grad_y(x,y))
    return np.array([[grad_x(x,y)], [grad_y(x, y)]])

def cost(w):
    x = w[0][0]
    y = w[1][0]
    return (x*x + y - 7)**2 + (x - y + 1)**2

def myGD(w):
    for it in range(500):
        print(np.linalg.norm(grad(w)))
        if np.linalg.norm(grad(w))<1e-3:
            break
        w -= grad(w)*eta
    
    return w, it

eta = 0.015
w = np.array([[0], [4]], np.float64)
w0, it = myGD(w)

print('Solutions: {} at x, y: {} in {} interators.'.format(cost(w0), w0, it))

w = np.array([[-5], [-12]], np.float64)
w0, it = myGD(w)

print('Solutions: {} at x, y: {} in {} interators.'.format(cost(w0), w0, it))
```
